```python
# Collatz Conjecture
# Collatz Sequence
n=int(input())
while(n!=1):
    if n%2==0:
        n=n//2
    elif n%2!=0:
        n=3*n+1
    print(n,end=' ')
```

```python
n=int(input())
s=(int(n**0.5))
i=2
c=0
for i in range(i,s+1):
    if n % i == 0:
        c=1
        break
if c==0:
    print("Prime")
else:
    print("Composite")
```
